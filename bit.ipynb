```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

height = 512
width = 512
```

```python
img = np.random.randint(2, size=(height, width, 1), dtype=np.uint8)
img1 = np.random.randint(256, size=(height, width, 1), dtype=np.uint8)
img2 = np.random.randint(65536, size=(height, width, 1), dtype=np.uint16)
img3 = np.random.randint(16777216, size=(height, width, 1), dtype=np.uint32)

plt.figure(figsize=(8, 8))
plt.subplot(2, 2, 1), plt.imshow(img, cmap='gray')
plt.title("2-bit")
plt.subplot(2, 2, 2), plt.imshow(img1, cmap='gray')
plt.title("8-bit")
plt.subplot(2, 2, 3), plt.imshow(img2, cmap='gray')
plt.title("16-bit")
plt.subplot(2, 2, 4), plt.imshow(img3, cmap='gray')
plt.title("24-bit")
plt.show()
```
